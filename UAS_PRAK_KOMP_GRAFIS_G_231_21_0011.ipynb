```python
#Import library pandas dan matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('stock.csv')
data
```

```python
# A. Distribusi data pada setiap atribut purchase Price, current price dan gain/loss dengan menggunakan grafik histogram

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(data['Purchase Price'], bins=10, edgecolor='green')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
plt.hist(data['Current Price'], bins=10, edgecolor='red')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
plt.hist(data['Gain/Loss'], bins=10, edgecolor='yellow')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
# B. Membuat atribut dan simbol untuk 5 dividen tertinggi
top_dividend = data.nlargest(5, 'Dividend/share')[['Stock Name', 'Symbol', 'Dividend/share']]
print('Top 5 Stocks with highest Dividend/share:')
print(top_dividend)
```

```python
# C. Membuat atribut dan simbol untuk 5 market value tertinggi

top_market_value = data.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
print('\nTop 5 Stocks with highest Market Value:')
print(top_market_value)
```

```python
# D. Membuat atribut dan simbol untuk 7 annual yields tertinggi

top_annual_yield = data.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]
print('\nTop 7 Stocks with highest Annual Yield:')
print(top_annual_yield)
```
